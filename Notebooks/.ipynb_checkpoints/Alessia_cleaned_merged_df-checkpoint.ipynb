```python
#Is the completion rate (number of customers who reached the next step out of the total in a given step) for the test group the same as the control group?

#Filtering the last start occurence
start_last_test =  test_df[test_df['process_step'] == 'start'].groupby('client_id').last()

# Filtering the last confirm occurence
confirm_last_test =  test_df[test_df['process_step'] == 'confirm'].groupby('client_id').last()

# List of clients who just started
clients_started_test = set(start_last_test.index)

# List of client who finished the process
clients_confirmed_test = set(confirm_last_test.index)

# Nb of client who started
total_clients_started_test = len(clients_started_test)

# Nb of client who finished the process
total_clients_confirmed_test = len(clients_confirmed_test)

# Completion Rate
completion_rate_test = total_clients_confirmed_test / total_clients_started_test if total_clients_started_test > 0 else 0

# Completion rate in % 
print(f"Completion rate : {completion_rate_test:.2%}")
```

```python
#Is the completion rate (number of customers who reached the next step out of the total in a given step) for the test group the same as the control group?

#Filtering the last start occurence
start_last_control =  control_df[control_df['process_step'] == 'start'].groupby('client_id').last()

# Filtering the last confirm occurence
confirm_last_control =  control_df[control_df['process_step'] == 'confirm'].groupby('client_id').last()

# List of clients who just started
clients_started_control = set(start_last_control.index)

# List of client who finished the process
clients_confirmed_control = set(confirm_last_control.index)

# Nb of client who started
total_clients_started_control = len(clients_started_control)

# Nb of client who finished the process
total_clients_confirmed_control = len(clients_confirmed_control)

# Completion Rate
completion_rate_control = total_clients_confirmed_control / total_clients_started_control if total_clients_started_control > 0 else 0

# Completion rate in % 
print(f"Completion rate : {completion_rate_control:.2%}")
```

```python
test_df["completion_rate"] = total_clients_confirmed_test / total_clients_started_test if total_clients_started_test > 0 else 0
control_df["completion_rate"] = total_clients_confirmed_control / total_clients_started_control if total_clients_started_control > 0 else 0
```

```python
#proportion test
import numpy as np
import statsmodels.api as sm

# Number of clients who started and confirmed for test group
num_confirmed_test = total_clients_confirmed_test
num_started_test = total_clients_started_test

# Number of clients who started and confirmed for control group
num_confirmed_control = total_clients_confirmed_control
num_started_control = total_clients_started_control

# Create arrays for the count of successes and the number of observations
count = np.array([num_confirmed_test, num_confirmed_control])
nobs = np.array([num_started_test, num_started_control])

# Perform the z-test for proportions
z_stat, p_value = sm.stats.proportions_ztest(count, nobs)

# Adjust for one-tailed test
if z_stat > 0:
    p_value_one_tailed = p_value / 2
else:
    p_value_one_tailed = 1 - (p_value / 2)

# Print the results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value (one-tailed): {p_value_one_tailed:.4f}")

# Check significance at alpha = 0.05
alpha = 0.05
if p_value_one_tailed < alpha:
    print("The completion rate for the test group is significantly higher than the control group.")
else:
    print("The completion rate for the test group is not significantly higher than the control group.")
```

```python
#Is the completion rate (number of customers who reached the next step out of the total in a given step) for the test group the same as the control group?

#Measuring stat significance with 2samples T test
import matplotlib.pyplot as plt
import scipy.stats as st

# Create arrays of completion status (1 for completed, 0 for not completed)
# For each client, check if they confirmed
completion_status_test = np.array([1 if client in clients_confirmed_test else 0 for client in clients_started_test])
completion_status_control = np.array([1 if client in clients_confirmed_control else 0 for client in clients_started_control])

# Perform two-sample t-test
t_stat, p_value = st.ttest_ind(completion_status_test, completion_status_control, equal_var=False, alternative="greater")

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Check significance at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("The completion rate for the test group is significantly greater than the control group.")
else:
    print("The completion rate for the test group is not significantly greater than the control group.")
```

```python
import pandas as pd

# Function to calculate and clean completion times
def calculate_and_clean_completion_times(df, max_conversion_time_hours=1):
    # Get the latest 'start' time and earliest 'confirm' time for each client
    start_times = df[df['process_step'] == 'start'].groupby('client_id')['date_time'].last()
    confirm_times = df[df['process_step'] == 'confirm'].groupby('client_id')['date_time'].first()

    # Merge start and confirm times into a DataFrame
    merged = pd.DataFrame({
        'start_time': start_times,
        'confirm_time': confirm_times
    })

    # Convert to datetime
    merged['start_time'] = pd.to_datetime(merged['start_time'], errors='coerce')
    merged['confirm_time'] = pd.to_datetime(merged['confirm_time'], errors='coerce')

    # Drop rows with missing start_time or confirm_time
    merged = merged.dropna(subset=['start_time', 'confirm_time'])

    # Calculate completion time in minutes
    merged['completion_time'] = (merged['confirm_time'] - merged['start_time']).dt.total_seconds() / 60

    # Drop negative completion times
    merged = merged[merged['completion_time'] >= 0]

    # Filter out clients who took longer than max_conversion_time_hours
    merged = merged[merged['completion_time'] <= max_conversion_time_hours * 60]

    return merged

# Calculate and clean completion times for test and control groups
completion_times_test = calculate_and_clean_completion_times(test_df)
completion_times_control = calculate_and_clean_completion_times(control_df)

# Merge the calculated completion times back into the original DataFrames
test_df = test_df.merge(completion_times_test[['completion_time']], left_on='client_id', right_index=True, how='left')
control_df = control_df.merge(completion_times_control[['completion_time']], left_on='client_id', right_index=True, how='left')

# Drop NaN values from the 'completion_time' column in both DataFrames
test_df = test_df.dropna(subset=['completion_time'])
control_df = control_df.dropna(subset=['completion_time'])

# Print data distributions to debug
print("Test Group - Completion Time Distribution:")
print(test_df['completion_time'].describe())
print("\nControl Group - Completion Time Distribution:")
print(control_df['completion_time'].describe())

# Calculate and print statistics for test and control groups
test_mean = test_df['completion_time'].mean()
test_std = test_df['completion_time'].std()
control_mean = control_df['completion_time'].mean()
control_std = control_df['completion_time'].std()

print("\nTest Group - Mean:", test_mean)
print("Test Group - Std:", test_std)

print("Control Group - Mean:", control_mean)
print("Control Group - Std:", control_std)
```

```python
# Perform two-sample t-test
t_stat, p_value = st.ttest_ind(test_df['completion_time'].dropna(), control_df['completion_time'].dropna(), equal_var=False, alternative='less')

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Check significance at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("The mean completion time for the test group is significantly shorter than the control group.")
else:
    print("The mean completion time for the test group is significantly higher than the control group.")
```

```python
# Function to calculate error rate (proportion of clients who revisited a previous step)
def calculate_error_data(df):
    # Create a DataFrame with steps and their occurrence order for each client
    steps_df = df[['client_id', 'process_step']].copy()
    steps_df['step_order'] = steps_df.groupby('client_id').cumcount()

    # Merge the steps DataFrame with itself to find revisits
    merged = pd.merge(steps_df, steps_df, on='client_id', suffixes=('_first', '_second'))
    revisits = merged[(merged['step_order_second'] < merged['step_order_first']) & (merged['process_step_first'] == merged['process_step_second'])]

    # Count unique clients with any revisit
    clients_with_revisits = revisits['client_id'].unique()
    total_clients = df['client_id'].nunique()

    # Return the error count and total client count
    return len(clients_with_revisits), total_clients
```

```python
# Calculate error data for test and control groups
error_count_test, total_clients_test = calculate_error_data(test_df)
error_count_control, total_clients_control = calculate_error_data(control_df)

# Convert to error rates
error_rate_test = error_count_test / total_clients_test if total_clients_test > 0 else 0
error_rate_control = error_count_control / total_clients_control if total_clients_control > 0 else 0


# Prepare data for t-test using error counts and total counts
# Create error and non-error counts for both groups
error_data_test = [1] * error_count_test + [0] * (total_clients_test - error_count_test)
error_data_control = [1] * error_count_control + [0] * (total_clients_control - error_count_control)

# Perform two-sample t-test
t_stat, p_value = st.ttest_ind(error_data_test, error_data_control, equal_var=False, alternative='less')

# Print results
print(f"Error Rate Test Group: {error_rate_test:.2%}")
print(f"Error Rate Control Group: {error_rate_control:.2%}")

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Check significance at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("The error rate in the test group is significantly smaller than in the control group.")
else:
    print("The error rate in the test group is significantly higher than in the control group.")
```

```python
test_df["error_rate"] = error_count_test / total_clients_test if total_clients_test > 0 else 0
control_df["error_rate"] = error_count_control / total_clients_control if total_clients_control > 0 else 0
```

```python
#which gender takes longer in control and test?

# Filter out the 'U' gender
test_df = test_df[test_df['gendr'].isin(['M', 'F'])]
control_df = control_df[control_df['gendr'].isin(['M', 'F'])]

test_df.reset_index(drop=True, inplace=True)
control_df.reset_index(drop=True, inplace=True)

# Calculate mean completion times for males and females in the test group
test_males = test_df[test_df['gendr'] == 'M']['completion_time']
test_females = test_df[test_df['gendr'] == 'F']['completion_time']

# Calculate mean completion times for males and females in the control group
control_males = control_df[control_df['gendr'] == 'M']['completion_time']
control_females = control_df[control_df['gendr'] == 'F']['completion_time']

# Perform t-tests
t_stat_test_gender, p_value_test_gender = st.ttest_ind(test_males, test_females, equal_var=False, alternative = "less")
t_stat_control_gender, p_value_control_gender = st.ttest_ind(control_males, control_females, equal_var=False, alternative = "less")

# Print results
print("Test Group Mean Completion Time (Males):", test_males.mean())
print("Test Group Mean Completion Time (Females):", test_females.mean())
print("Control Group Mean Completion Time (Males):", control_males.mean())
print("Control Group Mean Completion Time (Females):", control_females.mean())

print("\nT-test for Gender in Test Group: T-statistic =", t_stat_test_gender, ", P-value =", p_value_test_gender)

alpha = 0.05
if p_value_test_gender < alpha:
    print("The average completion time for the females in the test group is significantly higher than the males")
else:
    print("The average completion time for the males in the test group is significantly higher than the females")
    
print("T-test for Gender in Control Group: T-statistic =", t_stat_control_gender, ", P-value =", p_value_control_gender)

alpha = 0.05
if p_value_control_gender < alpha:
    print("The average completion time for the females in the control group is significantly higher than the males")
else:
    print("The average completion time for the males in the control group is significantly higher than the females")
```

```python
# Define age bins and labels
age_bins = [17, 25, 35, 45, 55, 65, 75, 85, 95]
age_labels = ["17-25", "26-35", "36-45", "46-55", "56-65", "66-75", "76-85", "86-95"]

# Assign clients to age groups
test_df.loc[:, 'age_group'] = pd.cut(test_df['clnt_age'], bins=age_bins, labels=age_labels, right=False)
control_df.loc[:, 'age_group'] = pd.cut(control_df['clnt_age'], bins=age_bins, labels=age_labels, right=False)

test_df = test_df.dropna(subset=['age_group'])
control_df = control_df.dropna(subset=['age_group'])

# Calculate the mean completion time for each age group
mean_completion_times_test = test_df.groupby('age_group', observed=False)['completion_time'].mean()

mean_completion_times_control = control_df.groupby('age_group', observed=False)['completion_time'].mean()


# Print the mean completion times for each age group
print("\nMean Completion Times by Age Group (Test Group):")
print(mean_completion_times_test)

print("\nMean Completion Times by Age Group (Control Group):")
print(mean_completion_times_control)

# Perform ANOVA to compare completion times between age groups in the test and control groups
age_grouped_test = [test_df[test_df['age_group'] == age]['completion_time'] for age in age_labels]
age_grouped_control = [control_df[control_df['age_group'] == age]['completion_time'] for age in age_labels]

# Filter out empty groups
age_grouped_test = [group for group in age_grouped_test if not group.empty]
age_grouped_control = [group for group in age_grouped_control if not group.empty]

# Perform ANOVA
f_stat, p_value = st.f_oneway(*age_grouped_test, *age_grouped_control)

# Print ANOVA results
print("\nANOVA F-statistic for Age Groups:", f_stat)
print("ANOVA P-value for Age Groups:", p_value)
test_df.columns
```

```python
# Find the slowest gender in test and control
slowest_gender_test = test_df.groupby('gendr')['completion_time'].mean().idxmax()
slowest_gender_control = control_df.groupby('gendr')['completion_time'].mean().idxmax()

# Find the slowest age group in test and control
slowest_age_test = test_df.groupby('age_group')['completion_time'].mean().idxmax()
slowest_age_control = control_df.groupby('age_group')['completion_time'].mean().idxmax()

print("\nSlowest Gender in Test Group:", slowest_gender_test)
print("Slowest Gender in Control Group:", slowest_gender_control)

print("\nSlowest Age Group in Test Group:", slowest_age_test)
print("Slowest Age Group in Control Group:", slowest_age_control)

# Compare slowest gender between test and control
test_slowest_gender_times = test_df[test_df['gendr'] == slowest_gender_test]['completion_time']
control_slowest_gender_times = control_df[control_df['gendr'] == slowest_gender_control]['completion_time']

t_stat_gender, p_value_gender = st.ttest_ind(test_slowest_gender_times, control_slowest_gender_times, equal_var=False, alternative = "greater")
print("\nT-test for Slowest Gender: T-statistic =", t_stat_gender, ", P-value =", p_value_gender)

alpha = 0.05
if p_value < alpha:
    print("The average completion time for the females in the test group is significantly higher than the females in the control")
else:
    print("The average completion time for the males in the control group is significantly higher than the females in the test")


# Compare slowest age group between test and control
test_slowest_age_times = test_df[test_df['age_group'] == slowest_age_test]['completion_time']
control_slowest_age_times = control_df[control_df['age_group'] == slowest_age_control]['completion_time']

t_stat_age, p_value_age = st.ttest_ind(test_slowest_age_times, control_slowest_age_times, equal_var=False, alternative = "greater")
print("T-test for Slowest Age Group: T-statistic =", t_stat_age, ", P-value =", p_value_age)
alpha = 0.05
if p_value < alpha:
    print("The average completion time for clients 76-85 in the test group is significantly higher than in the control")
else:
    print("The average completion time for clients 76-85 in the control group is significantly higher than the in the test")
```
