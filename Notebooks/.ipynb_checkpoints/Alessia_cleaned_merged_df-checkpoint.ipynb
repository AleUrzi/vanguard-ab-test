```python
# Split the DataFrame into a test group
test_df = final_combined_df[final_combined_df['Variation'] == 'Test']

# Split the DataFrame into a control group
control_df = final_combined_df[final_combined_df['Variation'] == 'Control']
```

```python
#Is the completion rate (number of customers who reached the next step out of the total in a given step) for the test group the same as the control group?

#Filtering the last start occurence
start_last_test =  test_df[test_df['process_step'] == 'start'].groupby('client_id').last()

# Filtering the last confirm occurence
confirm_last_test =  test_df[test_df['process_step'] == 'confirm'].groupby('client_id').last()

# List of clients who just started
clients_started_test = set(start_last_test.index)

# List of client who finished the process
clients_confirmed_test = set(confirm_last_test.index)

# Nb of client who started
total_clients_started_test = len(clients_started_test)

# Nb of client who finished the process
total_clients_confirmed_test = len(clients_confirmed_test)

# Completion Rate
completion_rate_test = total_clients_confirmed_test / total_clients_started_test if total_clients_started_test > 0 else 0

# Completion rate in % 
print(f"Completion rate : {completion_rate_test:.2%}")
```

```python
#Is the completion rate (number of customers who reached the next step out of the total in a given step) for the test group the same as the control group?

#Filtering the last start occurence
start_last_control =  control_df[control_df['process_step'] == 'start'].groupby('client_id').last()

# Filtering the last confirm occurence
confirm_last_control =  control_df[control_df['process_step'] == 'confirm'].groupby('client_id').last()

# List of clients who just started
clients_started_control = set(start_last_control.index)

# List of client who finished the process
clients_confirmed_control = set(confirm_last_control.index)

# Nb of client who started
total_clients_started_control = len(clients_started_control)

# Nb of client who finished the process
total_clients_confirmed_control = len(clients_confirmed_control)

# Completion Rate
completion_rate_control = total_clients_confirmed_control / total_clients_started_control if total_clients_started_control > 0 else 0

# Completion rate in % 
print(f"Completion rate : {completion_rate_control:.2%}")
```

```python
#Measuring stat significance with 2samples T test
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

# Create arrays of completion status (1 for completed, 0 for not completed)
# For each client, check if they confirmed
completion_status_test = np.array([1 if client in clients_confirmed_test else 0 for client in clients_started_test])
completion_status_control = np.array([1 if client in clients_confirmed_control else 0 for client in clients_started_control])

# Perform two-sample t-test
t_stat, p_value = st.ttest_ind(completion_status_test, completion_status_control, equal_var=False, alternative="greater")

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Check significance at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("The completion rate for the test group is significantly greater than the control group.")
else:
    print("The completion rate for the test group is not significantly greater than the control group.")
```

```python
import pandas as pd
from scipy import stats

# Convert 'date_time' to datetime
test_df.loc[:, 'date_time'] = pd.to_datetime(test_df['date_time'])
control_df.loc[:, 'date_time'] = pd.to_datetime(control_df['date_time'])

# Function to calculate completion times for each client
def calculate_completion_times(df):
    # Get the max 'start' time and max 'confirm' time for each client
    start_times = df[df['process_step'] == 'start'].groupby('client_id')['date_time'].max()
    confirm_times = df[df['process_step'] == 'confirm'].groupby('client_id')['date_time'].max()

    # Merge start and confirm times into a DataFrame
    merged = pd.DataFrame({
        'start_time': start_times,
        'confirm_time': confirm_times
    })

    # Calculate completion time
    merged['completion_time'] = (merged['confirm_time'] - merged['start_time']).dt.total_seconds()

    # Drop rows where completion time is missing (incomplete processes)
    merged = merged.dropna(subset=['completion_time'])

    return merged['completion_time']

# Calculate completion times for test and control groups
completion_times_test = calculate_completion_times(test_df[test_df['client_id'].isin(clients_confirmed_test)])
completion_times_control = calculate_completion_times(control_df[control_df['client_id'].isin(clients_confirmed_control)])

# Calculate and print statistics for test and control groups
test_mean = completion_times_test.mean()
test_std = completion_times_test.std()
control_mean = completion_times_control.mean()
control_std = completion_times_control.std()

print("Test Group - Mean:", test_mean)
print("Test Group - Std:", test_std)

print("Control Group - Mean:", control_mean)
print("Control Group - Std:", control_std)
```

```python
from scipy import stats

# Perform a t-test

t_stat, p_value = stats.ttest_ind(completion_times_test, completion_times_control, equal_var=False, alternative="greater")

print(f"T-statistic: {t_stat}, P-value: {p_value}")
alpha = 0.05
if p_value < alpha:
    print("The average completion time for the test group is significantly greater than the control group.")
else:
    print("The average completion time for the test group is not significantly greater than the control group.")
```

```python

# Merge back the client demographics (age and gender) for further analysis
completion_times_test = pd.merge(completion_times_test, test_df[['client_id', 'clnt_age', 'gendr']].drop_duplicates(), on='client_id')
completion_times_control = pd.merge(completion_times_control, control_df[['client_id', 'clnt_age', 'gendr']].drop_duplicates(), on='client_id')

# Function to analyze by age and gender
def analyze_age_gender(df, group_name):
    # Define age bins (example: 18-24, 25-34, 35-44, 45-54, 55+)
    age_bins = [0, 24, 34, 44, 54, 100]
    age_labels = ['18-24', '25-34', '35-44', '45-54', '55+']
    
    # Add age group column
    df['age_group'] = pd.cut(df['clnt_age'], bins=age_bins, labels=age_labels, right=True)
    
    # Analyze by age and gender
    age_gender_stats = df.groupby(['age_group', 'gendr'])['completion_time'].agg(['mean', 'std', 'count']).reset_index()
    
    print(f"\n{group_name} Group - Age and Gender Analysis")
    print(age_gender_stats)

# Perform the age and gender analysis for both test and control groups
analyze_age_gender(completion_times_test, 'Test')
analyze_age_gender(completion_times_control, 'Control')

# 2-Sample T-Test to compare completion times between test and control groups
t_stat, p_value = stats.ttest_ind(completion_times_test['completion_time'], completion_times_control['completion_time'], equal_var=False)

print("\n2-Sample T-Test Results:")
print(f"T-statistic: {t_stat}, P-value: {p_value}")

print("\nTest Group - Mean:", completion_times_test['completion_time'].mean())
print("Test Group - Std:", completion_times_test['completion_time'].std())

print("\nControl Group - Mean:", completion_times_control['completion_time'].mean())
print("Control Group - Std:", completion_times_control['completion_time'].std())
```
